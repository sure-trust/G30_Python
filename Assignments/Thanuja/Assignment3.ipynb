```python
#1.Fiction game (Hangman)
list1=["Samsung","Apple","Lenovo","Panasonic","Xiaomi","Bajaj","Tata","Cisco","Dell","Asus","Nvidia","Intel"]

player1=input("Enter name of player1:")

player2=input("Enter name of player2:")

chose_word=input("Enter a word from list1 by player1 ensure all are in lowercase:")

turns=6

guesses=''

print(player2,"Guessing the words")

while turns>0:

  failed=0

  for char in chose_word:

    if char in guesses:

      print(char,end=" ")

    else:

      print(" _ ",end=" ")

      failed=failed+1

  print("\n")

  if failed==0:

    print(player2,"won")

    break

  guess=input("Guess a character:")

  g=guess.lower()

  guesses+=g

  if g not in chose_word:

    turns-=1

    print("Wrong .Please guess for another character!!!")

    print("You have still ",turns,"to guess the word.")

    if turns==0:

      print(player1,"won")


```

```python
#2.Grocery Budget Tracker

item=[]

price=[]

quantity=[]

budget=float(input("Enter the budget :"))

print("Enter 1.Add a item 2.Exit")

op=int(input("Enter your option : "))

am_left=budget

index=0

while op<=2 and op>0:

  print("You have amount : ",am_left)

  if op==1:

    name=input("Enter name of item :")

    weight=int(input("Enter the quantity of item:"))

    mrp=float(input("Enter mrp :"))

    am_item=mrp*weight

    if am_item<=am_left:

      item.insert(index,name)

      quantity.insert(index,weight)

      price.insert(index,mrp)

      index+=1

      am_left-=am_item

      print("Amount left after purchase : ",am_left)

    else:

      print("Amount to be paid for item",am_item)

      print("it is out of your budget:",am_left)

  elif op==2:

    break

  print("Enter 1.Add a item 2.Exit")

  op=int(input("Enter your option : "))

else:

  print("Invalid option")

print("Product name    quantity     price")

for i in range(len(price)):

  print(i+1," . ",item[i],quantity[i],price[i])

print("The bill of items :",budget - am_left)
```

```python
#3.Password Strength Checker

password=input("Enter your password : ")

special_characters=['!','#','@','$','%','^','&','*','(',')','-','+','.','?',':']

has_lower=False

has_upper=False

has_special_character=False

has_digit=False

if len(password) >=8:

  for i in range(len(password)):

    if password[i].islower():

      has_lower=True

    elif password[i].isupper():

      has_upper=True

    elif password[i].isdigit():

      has_digit=True

    elif password[i] in special_characters:

      has_special_character=True


else:

  print("The password length must be atleast 8")



if has_lower and has_upper and has_special_character and has_digit:

  print("Password Strength is very strong enough")

elif (has_lower or has_upper) and (has_special_character or has_digit):

  print("Password Strength is moderate")

  print("To build a strong password do the fllg changes")


  if not(has_lower):

    print("include a lower character")

  if not(has_upper):

    print("include a upper character")

  if not(has_special_character):

    print("Include a special character")

  if not(has_digit):

    print("include a digit")


else:

  print("The password strength is very weak.")

  print("To build a strong password do the fllg changes")


  if not(has_lower):

    print("include a lower character")

  if not(has_upper):

    print("include a upper character")

  if not(has_special_character):

    print("Include a special character")

  if not(has_digit):

    print("include a digit")


```

```python

#4.Personal budgetting app

income_name=[]

expenses_name=[]

expenses=[]

income=[]

print("Enter 1.Incomes 2.Expenses 3.Exit")

opt=int(input("enter the option :"))

index1=0

index2=0

total_incomes=0

total_expenses=0

while opt<=3 and opt>0:

  if opt==1:

    print("Enter income details : ")

    iname=input("Enter name of income : ")

    income_value=float(input("Enter value of income : "))

    income_name.insert(index1,iname)

    income.insert(index1,income_value)

    index1+=1

    total_incomes+=income_value

  if opt==2:

    print("Enter expense details : ")

    ename=input("Enter name of expense : ")

    expenses_value=float(input("Enter value of expense : "))

    expenses_name.insert(index2,ename)

    expenses.insert(index2,expenses_value)

    index2+=1

    total_expenses+=expenses_value

  if opt==3:

    break

  print("Enter 1.Incomes 2.Expenses 3.Exit")

  opt=int(input("enter the option :"))

else:
  print("Invalid option")


print("The expenses are :")

for i in range(len(expenses)):

    print("Expense name =",expenses_name[i])

    print("Expense value = ",expenses[i])

print("Total values of expenses = ",total_expenses)


print("The incomes are :")

for i in range(len(income)):

    print("Income name =",income_name[i])

    print("Income value = ",income[i])

print("Total values of incomes = ",total_incomes)


balances=total_incomes-total_expenses

if balances>0:

  print("You have surplus amount left after all your expenses = ")

  print("Savings amount = ",balances)

elif balances==0:

  print("You have left with zero balances.Your incomes and expenses are equal.")

  print("Savings = 0")

else:

  print("Your expenses are out of range reduce it and save money !!!")

  reduced_amt=total_expenses-total_incomes

  print("Your over spent amount = ",reduced_amt)

  exp_without_item=0

  for exp in expenses:

    if exp>=reduced_amt:

      exp_without_item=total_expenses - exp

      break

  print("After removing total expenses = ",exp_without_item)
```

```python
#5.quiz with feedback

questions=[]

options=[]

answers=[]

user_answers=[]

score=0

n=int(input("Enter number of questions for quiz : "))

j=0

k=0

for i in range(n):

  q=input("enter question :")

  questions.insert(j,q);

  opt=input("Enter you options : ")

  options.insert(j,opt)

  j=j+1

for i in range(n):

  ans=input("enter answers for the question")

  answers.insert(k,ans)

  k=k+1

player_name=input("enter player name :")

ch=input("Enter choice y/n :")


qno=0

while ch=='y' or ch=='Y':

  print("Here is your question : ",qno+1)

  print(questions[qno])

  print("Here is your options : ")

  print(options[qno])

  ans_chosen=input("enter your answer i.e option")

  if ans_chosen == answers[qno]:

    print("Your answer is correct : ")

    score+=1

    print("Your score is : ",score)

    qno+=1

  else:

    print("Your answer is wrong")

    print("The correct answer is : ",answers[qno])

    qno+=1


  ch=input("Enter choice y/n :")


percentage=(score/n)*100


if percentage >= 80:

        print("Great job! You really know your stuff.")

elif percentage >= 60:

        print("Not bad! Keep practicing to improve.")

else:

        print("Keep learning! You'll get better with practice.")



```

```python
#6.Currency converter

exchange_rates = {


        "INR": 74.5,
        "EUR": 0.85,
        "GBP": 0.72


}

amount_in_usd = float(input("Enter the amount in USD: "))
target_currency = input("Enter the target currency (INR, EUR, or GBP): ").upper()

if target_currency not in exchange_rates:
    print("Invalid currency code. Please enter INR, EUR, or GBP.")
else:
    converted_amount = amount_in_usd * exchange_rates[target_currency]
    print(f"{amount_in_usd:.2f} USD is approximately {converted_amount:.2f} {target_currency}")
```
