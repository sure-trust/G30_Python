```python
import os
import pickle

class Student:
    def __init__(self, id, name, age, course):
        self.id = id
        self.name = name
        self.age = age
        self.course = course

def add_student():
    try:
        id = input("Enter student ID: ")
        name = input("Enter student name: ")
        age = int(input("Enter student age: "))
        course = input("Enter student course: ")
        student = Student(id, name, age, course)
        students.append(student)
        save_data()
        print("Student added successfully.")
    except ValueError:
        print("Invalid input for age. Please enter a valid integer.")
    except Exception as e:
        print("An error occurred:", e)

def view_students():
    try:
        for student in students:
            print(f"ID: {student.id}, Name: {student.name}, Age: {student.age}, Course: {student.course}")
    except Exception as e:
        print("An error occurred:", e)

def update_student():
    try:
        id = input("Enter student ID to update: ")
        for student in students:
            if student.id == id:
                student.name = input("Enter updated name: ")
                student.age = int(input("Enter updated age: "))
                student.course = input("Enter updated course: ")
                save_data()
                print("Student record updated successfully.")
                return
        print("Student not found.")
    except ValueError:
        print("Invalid input for age. Please enter a valid integer.")
    except Exception as e:
        print("An error occurred:", e)

def delete_student():
    try:
        id = input("Enter student ID to delete: ")
        for student in students:
            if student.id == id:
                students.remove(student)
                save_data()
                print("Student record deleted successfully.")
                return
        print("Student not found.")
    except Exception as e:
        print("An error occurred:", e)

def search_student():
    try:
        search_criteria = input("Enter student name or course to search: ")
        found = False
        for student in students:
            if search_criteria.lower() in student.name.lower() or search_criteria.lower() in student.course.lower():
                print(f"ID: {student.id}, Name: {student.name}, Age: {student.age}, Course: {student.course}")
                found = True
        if not found:
            print("Student not found.")
    except Exception as e:
        print("An error occurred:", e)

def save_data():
    try:
        with open("students_data.pkl", "wb") as f:
            pickle.dump(students, f)
    except Exception as e:
        print("An error occurred while saving data:", e)

def load_data():
    try:
        if os.path.exists("students_data.pkl"):
            with open("students_data.pkl", "rb") as f:
                return pickle.load(f)
        else:
            return []
    except Exception as e:
        print("An error occurred while loading data:", e)
        return []

def menu():
    print("\n===== Student Management System =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Update Student")
    print("4. Delete Student")
    print("5. Search Student")
    print("6. Exit")

students = load_data()

while True:
    menu()
    choice = input("Enter your choice: ")

    if choice == '1':
        add_student()
    elif choice == '2':
        view_students()
    elif choice == '3':
        update_student()
    elif choice == '4':
        delete_student()
    elif choice == '5':
        search_student()
    elif choice == '6':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```
