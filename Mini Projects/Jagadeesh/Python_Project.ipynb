```python

def get_contact_info():
  name = input('Enter The Name:')
  phone_number = int(input('enter the phone number:'))
  return name,phone_number

def convert_phone_number(phone_number):
    try:
        return int(phone_number)
    except ValueError:
        print("Invalid phone number. Please enter a valid integer.")
        return None

def contact_exists(contact_name, contact_list):
    for contact in contact_list:
        if contact['name'] == contact_name:
            return True
    return False

def view_contacts_by_name(contact_list, name):
    for contact in contact_list:
        if contact['name'] == name:
            print(contact)

def search_contacts(contact_list, search_query):
    for contact in contact_list:
        if search_query in contact['name'] or search_query in contact['phone_number']:
            print(contact)

contacts = []

def add_contact(name, phone_number):
    contact = {'name': name, 'phone_number': phone_number}
    contacts.append(contact)

class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

def update_contact(contact_list, name, new_phone_number):
    for contact in contact_list:
        if contact.name == name:
            contact.phone_number = new_phone_number
            break

def delete_contact(contact_list, name):
    for contact in contact_list:
        if contact.name == name:
            contact_list.remove(contact)
            break

def save_contacts_to_file(file_name, contact_list):
    with open(file_name, 'w') as file:
        for contact in contact_list:
            file.write(f"{contact['name']},{contact['phone_number']}\n")

def load_contacts_from_file(file_name):
    try:
        with open(file_name, 'r') as file:
            for line in file:
                name, phone_number = line.strip().split(',')
                add_contact(name, phone_number)
    except FileNotFoundError:
        print("Contact file not found. Starting with an empty contact list.")

def main():
    load_contacts_from_file("contacts.txt")

    while True:
        print("\n1. Add Contact")
        print("2. View Contacts")
        print("3. Update Contact")
        print("4. Delete Contact")
        print("5. Search Contacts")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name, phone_number = get_contact_info()
            add_contact(name, phone_number)
        elif choice == '2':
            print(contacts)
        elif choice == '3':
            name = input("Enter the name of the contact to update: ")
            new_phone_number = input("Enter the new phone number: ")
            update_contact(contacts, name, new_phone_number)
        elif choice == '4':
            name = input("Enter the name of the contact to delete: ")
            delete_contact(contacts, name)
        elif choice == '5':
            search_query = input("Enter name or phone number to search: ")
            search_contacts(contacts, search_query)
        elif choice == '6':
            save_contacts_to_file("contacts.txt", contacts)
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
